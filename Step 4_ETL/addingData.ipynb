## This is a utility Script to divide the data from the the original IRIS dataset into different sources. Ignore this if you are following the tutorial

```python
import pandas as pd
import numpy as np


df = pd.read_csv("dataset/dataset.csv")
df.head()
```

```python
df["Diseases"] = np.random.choice(["Fungi","Bacteria","Virus"], size=len(df))
df["Cost"] = np.random.uniform(100,500, size=len(df)).round(2)
```

```python
df
```

```python
df.to_csv("dataset/dataset_add.csv", index=False)
```

```python
source4 = df[90:120]
json_data = source4.to_json(orient='records', indent=4)
print(json_data)
```

```python
source5 = df[120:]
json_data = source5.to_json(orient="records", indent=4)
print(json_data)
```
